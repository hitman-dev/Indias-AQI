#### Getting all the active AQI Stations in India

```python
import requests
import pandas as pd
```

Get your own API Key From [Here](https://aqicn.org/data-platform/token/)

```python
INDIAN_BOUNDS = "6.4626999,68.1097,35.513327,97.39535869999999"  # latitude,longitude boundaries of India

API_KEY = "{API_KEY}" # use your own API_KEY Here
boundaryUrl = f"https://api.waqi.info/map/bounds/?latlng={INDIAN_BOUNDS}&token={API_KEY}"
response = requests.get(boundaryUrl)
data = response.json()['data']
```

```python
lst = []
for item in data:
    if "India" in item.get("station").get("name"):
        lat = item.get("lat")
        lon = item.get("lon")
        station = item.get("station").get("name")
        dictionary = {
            'Station': station,
            'lat': lat,
            'lon': lon
        }
        lst.append(dictionary)
df = pd.DataFrame(lst)
df.head()
```

```python
df.to_csv('./AQI/Data/station_location.csv',index=False)
```
